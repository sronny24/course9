# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# your code goes here
table_name = 'master'
query = """
SELECT ConvertedComp FROM {}
""".format(table_name)

# Read the data into a DataFrame
df_data = pd.read_sql_query(query, conn)

plt.figure(figsize=(10, 6))
plt.hist(df_data['ConvertedComp'], bins=30, edgecolor='black')
plt.title('Histogram of ConvertedComp')
plt.xlabel('Converted Compensation')
plt.ylabel('Frequency')
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
table_name = 'master'
query = """
SELECT Age FROM {}
""".format(table_name)
df_data = pd.read_sql_query(query.format(table_name), conn)

# Plotting the box plot of Age
plt.figure(figsize=(10, 6))
sns.boxplot(df_data['Age'], color='skyblue')
plt.title('Box Plot of Age')
plt.xlabel('Age')
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
table_name = 'master'
query = """
SELECT Age, WorkWeekHrs FROM {}
""".format(table_name)
df_data = pd.read_sql_query(query.format(table_name), conn)

# Plotting the scatter plot of Age vs WorkWeekHrs
plt.figure(figsize=(10, 6))
plt.scatter(df_data['Age'], df_data['WorkWeekHrs'], alpha=0.5)
plt.title('Scatter Plot of Age vs WorkWeekHrs')
plt.xlabel('Age')
plt.ylabel('Work Week Hours')
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here

table_name = 'master'
query = """
SELECT WorkWeekHrs, CodeRevHrs, Age FROM {}
""".format(table_name)
df_data = pd.read_sql_query(query, conn)
plt.figure(figsize=(10, 8))
bubble = plt.scatter(df_data['WorkWeekHrs'], df_data['CodeRevHrs'], s=df_data['Age']*2, alpha=0.5, c='blue', edgecolors='grey', linewidth=2)
plt.title('Bubble Plot of WorkWeekHrs vs CodeRevHrs')
plt.xlabel('Work Week Hours')
plt.ylabel('Code Review Hours')
plt.show()
           
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
table_name = 'DatabaseWorkedWith,DatabaseDesireNextYear'
Query = """
SELECT COUNT(*) AS CountMySQLOnly
FROM (
    SELECT *
    FROM {}
    WHERE DatabaseWorkedWith = 'MySQL'
      AND DatabaseDesireNextYear NOT LIKE '%,%'  -- Exclure ceux qui mentionnent d'autres bases de données
)
""".format(table_name)
df = pd.read_sql_query(Query, conn)
df
```

```python
table_name = 'LanguageWorkedWith'
Query = """
SELECT LanguageWorkedWith, COUNT(*) as Count
FROM {}
WHERE LanguageWorkedWith IS NOT NULL
GROUP BY LanguageWorkedWith
ORDER BY Count DESC
LIMIT 30
""".format(table_name)
df = pd.read_sql_query(Query, conn)
df
```

```python
table_name = 'LanguageDesireNextYear'
Query = """
SELECT LanguageDesireNextYear, COUNT(*) as Count
FROM {}
WHERE LanguageDesireNextYear IS NOT NULL
GROUP BY LanguageDesireNextYear
ORDER BY Count DESC
LIMIT 30
""".format(table_name)
df = pd.read_sql_query(Query, conn)
df
```

```python
# your code goes here
table_name = 'DatabaseDesireNextYear'
Query = """
SELECT DatabaseDesireNextYear, COUNT(*) as Count
FROM {}
WHERE DatabaseDesireNextYear IS NOT NULL
GROUP BY DatabaseDesireNextYear
ORDER BY Count DESC
LIMIT 5
""".format(table_name)
df = pd.read_sql_query(Query, conn)

plt.figure(figsize=(8, 8))
plt.pie(df['Count'], labels=df['DatabaseDesireNextYear'], autopct='%1.1f%%', startangle=140)
plt.title('Top 5 Databases Respondents Wish to Learn Next Year')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.tight_layout()
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here

table_name = 'master'
Query = """
SELECT Age, 
       AVG(WorkWeekHrs) AS MedianWorkWeekHrs, 
       AVG(CodeRevHrs) AS MedianCodeRevHrs
FROM {}
WHERE Age >= 30 AND Age <= 35
GROUP BY Age
ORDER BY Age
""".format(table_name)
df = pd.read_sql_query(Query, conn)

plt.figure(figsize=(10, 6))
plt.bar(df['Age'], df['MedianWorkWeekHrs'], label='Median WorkWeekHrs', color='skyblue')
plt.bar(df['Age'], df['MedianCodeRevHrs'], bottom=df['MedianWorkWeekHrs'], label='Median CodeRevHrs', color='red')
plt.xlabel('Age')
plt.ylabel('Hours')
plt.title('Median WorkWeekHrs and CodeRevHrs for Age Group 30-35')
plt.legend()
plt.grid(True)
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
table_name = 'master'
Query = """
SELECT Age, AVG(ConvertedComp) AS MedianConvertedComp
FROM {}
WHERE Age >= 25 AND Age <= 30
GROUP BY Age
ORDER BY Age
""".format(table_name)
df = pd.read_sql_query(Query, conn)

plt.figure(figsize=(10, 6))
plt.plot(df['Age'], df['MedianConvertedComp'], marker='o', linestyle='-', color='b', label='Median ConvertedComp')
plt.xlabel('Age')
plt.ylabel('Median Converted Compensation')
plt.title('Median ConvertedComp for Ages 25-30')
plt.show()
```

```python
# your code goes here
table_name = 'master'
Query = """
SELECT Age, AVG(ConvertedComp) AS MedianConvertedComp
FROM {}
WHERE Age >= 45 AND Age <= 60
GROUP BY Age
ORDER BY Age
""".format(table_name)
df = pd.read_sql_query(Query, conn)

plt.figure(figsize=(10, 6))
plt.plot(df['Age'], df['MedianConvertedComp'], marker='o', linestyle='-', color='b', label='Median ConvertedComp')
plt.xlabel('Age')
plt.ylabel('Median Converted Compensation')
plt.title('Median ConvertedComp for Ages 45-60')
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python

table_name = 'DevType'
Query = """
SELECT DevType, COUNT(*) AS Count
FROM {}
GROUP BY DevType
ORDER BY Count DESC
""".format(table_name)
df = pd.read_sql_query(Query, conn)
df
```

```python
# your code goes here
table_name = 'master'
Query = """
SELECT MainBranch, COUNT(*) AS Count
FROM {}
GROUP BY MainBranch
ORDER BY Count DESC
""".format(table_name)
df = pd.read_sql_query(Query, conn)

plt.figure(figsize=(10, 6))
plt.barh(df['MainBranch'], df['Count'], color='skyblue') #barh is to create a horizontal bar chart
plt.xlabel('Count')
plt.ylabel('Main Branch')
plt.title('Count of Respondents by Main Branch')
plt.show()
```
